```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets,transforms

train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p = 0.4),
    transforms.RandomVerticalFlip(p = 0.4),
    #transforms.RandomRotation(degrees = [-15,15]),
    transforms.RandomAffine(degrees = [-20,20]),
    transforms.ToTensor(),
    transforms.Normalize((0.3107,),(0.3081,))
])
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.3107,),(0.3081,))
])

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.3107,),(0.3081,))
])
train_data = datasets.MNIST("../data",train = True,download = True,transform = train_transform)
test_data = datasets.MNIST("../data",train = False,transform = test_transform)
```

```python
BATCH_SIZE = 20
train_loader = torch.utils.data.DataLoader(train_data,batch_size = BATCH_SIZE,shuffle = True)
test_loader = torch.utils.data.DataLoader(test_data,batch_size = BATCH_SIZE,shuffle = False)
```

```python
for data,label in train_loader:
    print(data.shape)
    print(label)
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,32,kernel_size = 3,stride = 1)
        self.conv2 = nn.Conv2d(32,64,kernel_size = 2,stride = 1)
        self.conv3 = nn.Conv2d(64,128,kernel_size =3,stride =1)
        self.fc1 = nn.Linear(512,128)
        self.fc2 = nn.Linear(128,10)
        self.dropout1 = nn.Dropout2d(0.25)
        
    def forward(self,x):
        x = self.conv1(x)
        x = F.relu(x)
        x = F.max_pool2d(x,2)
        x = self.conv2(x)
        x = F.relu(x)
        x = F.max_pool2d(x,2)
        x = self.conv3(x)
        x = F.relu(x)
        x = F.max_pool2d(x,2)
        x = self.dropout1(x)
        x = torch.flatten(x,1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        output = x
        return output
```

```python
model = Net()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
```

```python
record_loss_train = []
record_loss_test = []

for epoch in range(1,11):
    for batch,(data,target) in enumerate(train_loader):
        optimizer.zero_grad()
        y_predict = model(data)
        loss = criterion(y_predict,target)
        loss.backward()
        optimizer.step()
        
        if batch % 500 == 0:
            print(f'#{epoch},data:{batch*20},loss: {loss}')
```

```python
_,predicted = torch.max(y_predict,1)
print(predicted)
print(target)
```

```python
#print(y_predict)
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data in test_loader:
        inputs,labels = data
        outputs = model(inputs)
        _,predicted = torch.max(outputs,1)
        total += len(outputs)
        correct += (predicted == labels).sum().item()

print(f"correct:{correct},accuracy:{correct}/{total} = {correct/total}")
```
